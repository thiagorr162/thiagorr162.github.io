```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score

import pandas as pd
```

# Validação cruzada

```python
df = pd.read_csv("iris.csv")
```

```python
df.head()
```

Primeiro criamos o conjunto y formado pela saída que queremos prever. E o conjunto X formado pelas covariadas que serão usadas na previsão.

```python
y = df["variety"]
X = df.drop(columns=["variety"])
```

Agora vamos separar o conjunto em treino e teste.

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=123)
```

```python
print("Tamanho de X_train:", X_train.shape)
print("Tamanho de X_test:", X_test.shape)
print("Tamanho de Y_train:", y_train.shape)
print("Tamanho de Y_test:", y_test.shape)
```

Agora vamos usar o conjunto de treino para treinar um modelo de Floresta Aleatória

```python
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)
```

Primeiro vamos calcular o erro do nosso algorítmo no conjunto de treino.

```python
train_pred = model.predict(X_train)
results_df = pd.DataFrame({'True Labels': y_train, 'Predicted Labels': train_pred})

results_df.head()
```

```python
train_accuracy = accuracy_score(y_train,train_pred)
print("Acurácia no conjunto de treino:",train_accuracy)
```

Agora vamos fazer o mesmo no conjunto de teste.

```python
test_pred = model.predict(X_test)
results_df = pd.DataFrame({'True Labels': y_test, 'Predicted Labels': test_pred})

results_df.head()
```

```python
test_accuracy = accuracy_score(y_test,test_pred)
print("Acurácia no conjunto de teste:",test_accuracy)
```

Note que a acurácia no teste foi menor que no treino, como esperado.
